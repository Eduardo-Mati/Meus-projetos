```python
# Importar as principais bibliotecas de ciência de dados para python.
import pandas as pd # planilha eletrônica e todas as suas operações.
import seaborn as sns # componentes de visualização de dados no formato html.
import matplotlib.pyplot as plt # biblioteca de visualização 2D e 3D para ciência de dados.
```

```python
# Criando um novo dataframe para coletar os dados e analisar.

df = pd.read_excel("dataset.xlsx")
```

```python
# Visualizando as primeiras 10 linhas do dataframe
df.head(10)
```

```python
# Visualizando as ultimas 10 linha do dataframe
df.tail(15)
```

```python
# Neste caso podemos dizer que existe uma CORRELAÇÃO FORTE POSITIVA entre essas variáveis
# Na medida que a variável X aumenta a variável Y também aumenta
sns.lmplot(x="Ano", y = "Cilindradas", data = df);
```

```python
sns.lmplot( x = "Ano", y = "ValorTabela", data = df);
```

```python
df = df.drop(columns=["Modelo"])
```

```python
df.corr()
```

```python
# Através do heatmap podemos observar a correlção entre todas as variáveis.
plt.rcParams["figure.figsize"] = (10, 8)
ax = sns.heatmap(df.corr(), annot=True)
```
