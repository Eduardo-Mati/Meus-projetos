```python
import tabula
import PyPDF2
```

```python
# Primeiro vamos ler o arquivo PDF que possui imagem de tabelas
# Carregando o arquivo PDF
lista_tabelas = tabula.read_pdf("arquivo2.pdf", pages="all")
```

```python
# Exibindo quantaas páginas foram carregadas
print(len(lista_tabelas))
```

```python
# Gerando um loop FOR para exibir cada página do PDF

for tabela in lista_tabelas:
    display(tabela)
```

```python
# Obtendo somente a página 1. Lembre-se que no Python o índice começa em zero, por isso passamos o parametro zero abaixo
tabela1 = lista_tabelas[0]
tabela1
```

```python
# Obtendo somente a página 2. Lembre-se que no Python o índice começa em zero, por isso passamos o parametro zero abaixo
tabela2 = lista_tabelas[1]
tabela2
```

```python
# Obtendo somente a página 3. Lembre-se que no Python o índice começa em zero, por isso passamos o parametro zero abaixo
tabela3 = lista_tabelas[2]
tabela3
```

```python
tabela3.info()
```

```python
tabela1.info()
```

```python
# Separar a variação percentual
# criar coluna 0 e coluna 1 e os valores que serão colocados detro será 1sem e 2sem
# Faz um SPLIT para separar no espaço, o comando STR é devido estarmos fazendo um tratamento em uma string
# comando EXPAND separa e cria colunas novas
tabela3[[0,1]] = tabela3['VARIACAO_PERCENTUAL'].str.split(" ", expand = True)
tabela3
```

```python
# Vamos excluir a coluna original com o comando DROP e utilizar o AXIS 1 (EIXO 1)
# Axis 1 = Colunas e Axis 2 = Linhas
tabela3 = tabela3.drop('VARIACAO_PERCENTUAL', axis = 1)
tabela3
```

```python
# Vamos renomear as colunas do nosso Dataframe
tabela3.columns = ['DESCRICAO', '1 SEMESTRE', '2 SEMESTRE', 'VAR_PERC_1_SEM','VAR_PERC_2_SEM']
tabela3
```

```python
tabela3.to_pdf('tabela3.pdf', index=False)
```

```python
from fpdf import FPDF
from tabulate import tabulate

# Convert the DataFrame to a string table
table_string = tabulate(tabela3, headers='keys', tablefmt='pdf')

# Create a PDF object
pdf = FPDF()
pdf.add_page()

# Set font
pdf.set_font("Arial", size=12)

# Add the table string to the PDF
pdf.multi_cell(0, 10, table_string)

# Save the PDF
pdf_output_path = "tabela3.pdf"
pdf.output(pdf_output_path)

print(f"DataFrame saved to {pdf_output_path}")
```
